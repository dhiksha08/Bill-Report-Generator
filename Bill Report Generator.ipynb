```python
from tkinter import *
import mysql.connector
import tkinter.messagebox
import random
import csv

root=Tk()
root.title("BILL DETAILS")
root.configure(background="white")
root.geometry("1200x900")

header1=Label(root,text="DDK JEWELLERS-SALEM",
padx=10,pady=10,font="Times 23 bold")
header1.grid(row=1,column=0,columnspan=7,padx=10,pady=10)

p=r=Final_price=0

p=StringVar(root)
p.set("Choose")
r=StringVar(root)
r.set("Choose")

l1=Label(root,text="Customer Name",padx=5,pady=5,font='Times 14')
l2=Label(root,text="Ornament Type",padx=5,pady=5,font='Times 14')
l3=Label(root,text="Product Code",padx=5,pady=5,font='Times 14')
l4=Label(root,text="Ornament Weight(in Gm.)",padx=5,pady=5,font='Times 14')

l12=Label(root,text="Purchase date(YYYY-MM-YY)",padx=5,pady=5,font='Times 14')

l5=Label(root,text="Discount(%)",padx=5,pady=5,font='Times 14')
l6=Label(root,text="Wastage(in Gm.)",padx=5,pady=5,font='Times 14')
L7=Label(root,text="Gold Purity(%)",padx=5,pady=5,font=("bold",14))
l8=Label(root,text="Cost of 1g GOLD(in Rs.)",padx=5,pady=5,font='Times 14')
l9=Label(root,text="Name of the Sales Person",padx=5,pady=5,font='Times 14')
l10=Label(root,text="Final Price(in Rs.:",padx=5,pady=5,font='Times 16')
#l11=Label(root,text=Final_price,padx=5,pady=5,font='Times 21')

L7.grid(row=7,column=0,padx=5,pady=5,sticky='w')
i=IntVar()
r1=Radiobutton(root,text="99.9%(24K)",value=1,variable=i,font=("bold",14))
r2=Radiobutton(root,text="91.7%(22K)",value=2,variable=i,font=("bold",14))
r1.grid(row=7,column=1,padx=1,pady=1,sticky='w')
r2.grid(row=7,column=2,padx=1,pady=1,sticky='w')

e1=Entry(root,width=20,borderwidth=3,font='Times 14')
e2=OptionMenu(root,p,"Toe Ring","Hipbelt","Armlets","Anklet","NoseStud","Earrings","Necklace")
e3=OptionMenu(root,r,"101","102","504","343","564","900","1020")
e4=Entry(root,width=20,borderwidth=3,font='Times 14')
e5=Entry(root,width=20,borderwidth=3,font='Times 14')

e6=Entry(root,width=20,borderwidth=3,font='Times 14')

e8=Entry(root,width=20,borderwidth=3,font='Times 14')
e9=Entry(root,width=20,borderwidth=3,font='Times 14')
#e10=Entry(root,width=20,borderwidth=3,font='Times 14')
e12=Entry(root,width=20,borderwidth=3,font='Times 14')

l1.grid(row=2,column=0,padx=10,pady=10,sticky='w')
e1.grid(row=2,column=1,padx=10,pady=10,sticky='w')
l2.grid(row=3,column=0,padx=10,pady=10,sticky='w')
e2.grid(row=3,column=1,padx=10,pady=10,sticky='w')
l3.grid(row=3,column=2,padx=10,pady=10,sticky='w')
e3.grid(row=3,column=3,padx=10,pady=10,sticky='w')
l4.grid(row=4,column=0,padx=10,pady=10,sticky='w')
e4.grid(row=4,column=1,padx=10,pady=10,sticky='w')
l5.grid(row=5,column=0,padx=10,pady=10,sticky='w')
e5.grid(row=5,column=1,padx=10,pady=10,sticky='w')
l6.grid(row=5,column=2,padx=10,pady=10,sticky='w')
e6.grid(row=5,column=3,padx=10,pady=10,sticky='w')
l8.grid(row=6,column=0,padx=10,pady=10,sticky='w')
e8.grid(row=6,column=1,padx=10,pady=10,sticky='w')

l9.grid(row=6,column=2,padx=10,pady=10,sticky='w')
e9.grid(row=6,column=3,padx=10,pady=10,sticky='w')
l10.grid(row=9,column=0,padx=10,pady=10,sticky='w')
#l11.grid(row=9,column=1,padx=10,pady=10,sticky='w')
#e10.grid(row=11,column=1,padx=10,pady=10,sticky='w')

l12.grid(row=4,column=2,padx=10,pady=10,sticky='w')
e12.grid(row=4,column=3,padx=10,pady=10,sticky='w')

def done():
    Customer_Name=str(e1.get())
    Ornament_Type=p.get()
    Product_Code=int(r.get())
    O_Weight=float(e4.get())
    Discount=float(e5.get())
    Wastage=float(e6.get())
    Pdate=e12.get()
    if i.get()==1:
        Purity=99.9
    else:
        Purity=91.7
    
    Cost_Of_1g_Gold=int(e8.get())
    Sale_attendee_Name=str(e9.get())
    #Sales_id=int(e10.get())
    
    calc1=O_Weight*Cost_Of_1g_Gold
    
    w=Wastage/100*calc1
    
    D=(calc1+w)*Discount/100
    
    Final_price=(calc1+w)-D
    
    l11=Label(root,text=Final_price,padx=5,pady=5,font='Times 21')
    l11.grid(row=9,column=1,padx=10,pady=10,sticky='w')

    #fill in your details in the respective parameters
    conn=mysql.connector.connect(host="",user="",passwd="",database="")
    cursor=conn.cursor()
    
    cursor.execute("""select Sales_id from ORNAMENT_PURCHASE_LIST;""")
    x=cursor.fetchall()
    
    while True:
        Sales_id=random.randint(1000,9999)
        if Sales_id not in(x):
            break

    sq1="""insert into ORNAMENT_PURCHASE_LIST
    values("%s","%s","%s","%s","%s","%s","%s","%s","%s","%s","%s","%s")"""

    t1=(Customer_Name,Ornament_Type,Product_Code,O_Weight,Discount,Wastage,Purity,Cost_Of_1g_Gold,Sale_attendee_Name,Sales_id,Final_price,Pdate)

    cursor.execute(sq1,t1)
    cursor.execute("commit")

    tkinter.messagebox.showinfo("DATABASE STATUS","Bill Record has been inserted......")
    e1.delete(0,'end')
    e4.delete(0,'end')
    e5.delete(0,'end')
    e6.delete(0,'end')

    e8.delete(0,'end')
    e9.delete(0,'end')
        #e12.delete(0,'end')

def display():
    root2=Tk()
    root2.title("Sales Report")
    root2.configure(background="grey")
    root2.geometry("900x900")
    #fill in your details in the respective parameters
    conn=mysql.connector.connect(host="",user="",passwd="",database="")
    cursor=conn.cursor()
    cursor.execute("select * from ORNAMENT_PURCHASE_LIST")
    result=cursor.fetchall()
    for index,x in enumerate(result):
        num=0
        for y in x:
            L18=Label(root2,text=y)
            L18.grid(row=index,column=num)
            num+=1
    csv_button=Button(root2,text="Save to Excel",command=lambda:write_to_csv(result))
    csv_button.grid(row=15,column=0)

def write_to_csv(result):
    fo=open("report.csv","w")
    wo=csv.writer(fo)
    for i in result:
        wo.writerow([i[0],i[1],i[2],i[3],i[4],i[5],i[6],i[7],i[8],i[9],i[10],i[11]])
    fo.close()
    
Done=Button(root,text="Done",height=2,width=18,font=("Italic",10),bg="blue",fg="white", command=done)
Done.grid(row=8,column=0,padx=10,pady=10)
    
display1=Button(root,text="Report",height=2,width=18,font=("Italic",10),bg="blue",fg="white", command=display)
display1.grid(row=8,column=2,padx=10,pady=10)

root.mainloop()
```
